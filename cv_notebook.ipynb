```python
import pandas as pd

confirmed_csv=pd.read_csv("../project_3/rawData/monthly_confirmed.csv")
deaths_csv=pd.read_csv("../project_3/rawData/monthly_deaths.csv")
recovery_csv=pd.read_csv("../project_3/rawData/monthly_recovery.csv")
```

```python
confirmed_df=pd.DataFrame(confirmed_csv)
deaths_df=pd.DataFrame(deaths_csv)
recovery_df=pd.DataFrame(recovery_csv)
```

```python
confirmed_df.head(15)
```

```python
# Change column names to replace "/" with "_"
new_columns = {}
for column in confirmed_df.columns:
    new_columns[column] = column.replace("/", "_")

confirmed_df.rename(columns=new_columns, inplace=True)

confirmed_df.head()
```

```python
# Change column names to replace "/" with "_"
new_columns = {}
for column in deaths_df.columns:
    new_columns[column] = column.replace("/", "_")

deaths_df.rename(columns=new_columns, inplace=True)

deaths_df.head()
```

```python
# Change column names to replace "/" with "_"
new_columns = {}
for column in recovery_df.columns:
    new_columns[column] = column.replace("/", "_")

recovery_df.rename(columns=new_columns, inplace=True)

recovery_df.head()
```
