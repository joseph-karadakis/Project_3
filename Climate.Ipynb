```python
import pandas as pd
import numpy as np
```

```python
# read in climate data csv
climate_data_raw = "Climate_Data.csv"

climate_df = pd.read_csv(climate_data_raw)
climate_df.head()
```

```python
# Chaning the column names
climate_df = climate_df.rename(columns={"STATION":"Station",
                                        "DATE":"Date",
                                        "LATITUDE": "Lat",
                                        "LONGITUDE": "Lon",
                                        "AIR_TEMP": "Air_temp",
                                        "IND_FOR_PRECIP": "Prcp",
                                        "IND_FOR_TEMP": "Temp",
                                        "SEA_SURF_TEMP": "Sea_Temp"})
climate_df.head(10)
```

```python
climate_df.dtypes
```

```python
# fill null values
climate_df_clean = climate_df.fillna(value=0)
climate_df_clean.head(10)
```

```python
climate_df_clean.count()
```

```python
climate_date = climate_df_clean.sort_values("Date")
climate_date
```

```python
climate_date = climate_date.copy()
climate_date = climate_date.astype({"Date":"datetime64"})
climate_date
```
